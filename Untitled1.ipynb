```python
import pandas as pd
import numpy as np

# Set seed
np.random.seed(42)

# Define global date range
dates = pd.date_range(start="2025-05-01", end="2025-10-31", freq="D")
n_rows = 10000

# Product definitions
product_to_type = {
    'ball': 'toy',
    'soldier': 'toy',
    'hand towel': 'towel',
    'bath towel': 'towel'
}
product_list = list(product_to_type.keys())

# === Helper: Parabolic units with zero cutoff ===
def parabolic_units(date, peak_date, spread_days=90, max_units=20):
    days_from_peak = (date - peak_date).days
    x = days_from_peak / spread_days
    y = -x**2 + 1  # Parabola: peak at 0, drops to 0 at ±1
    noise = np.random.normal(0, 0.15)
    value = max_units * (y + noise)
    return int(round(value)) if value > 0 else 0

# === Generate production data (peak Aug 15) ===
prod_dates = np.random.choice(dates, size=n_rows)
prod_units = [parabolic_units(d, pd.Timestamp("2025-08-15")) for d in prod_dates]

prod_df = pd.DataFrame({
    'date': prod_dates,
    'units': prod_units,
    'type': 'produced',
    'product': np.random.choice(product_list, size=n_rows)
})
prod_df['product_type'] = prod_df['product'].map(product_to_type)

# === Generate shipment data (peak Oct 20) ===
ship_dates = np.random.choice(dates, size=n_rows)
ship_units = [parabolic_units(d, pd.Timestamp("2025-10-20")) for d in ship_dates]

ship_df = pd.DataFrame({
    'date': ship_dates,
    'units': ship_units,
    'type': 'shipped',
    'product': np.random.choice(product_list, size=n_rows)
})
ship_df['product_type'] = ship_df['product'].map(product_to_type)

# === Combine and clean ===
df = pd.concat([prod_df, ship_df], ignore_index=True)
df = df[df['date'] <= pd.Timestamp("2025-10-31")]
df = df[df['units'] > 0].copy()
df['day'] = df['date'].dt.day
df = df.sort_values('date').reset_index(drop=True)
# After filtering zero units but before sorting:
df.loc[df['type'] == 'produced', 'units'] = (df.loc[df['type'] == 'produced', 'units'] * 0.8).round().astype(int)


# ✅ Ready for analysis or visualization
print(df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.nonparametric.smoothers_lowess import lowess

def stream_chart(
    data,
    x,
    y,
    units_col,
    hue=None,
    split=True,
    palette=None,
    global_scale=True,
    ax=None,
    smoothing_method="none",  # "none", "rolling", or "lowess"
    smoothing_window=3       # window size for rolling or frac for lowess
):
    """
    Stream chart with options for split/combined streams and smoothing.

    Parameters
    ----------
    data : pd.DataFrame
        Input data containing columns x, y, units_col, and optional hue.
    x : str
        Datetime column name.
    y : str
        Categorical column name (e.g., product_type).
    units_col : str
        Column name for stream width values.
    hue : str or None
        Binary split column name (e.g., 'produced' vs 'shipped').
    split : bool
        If True, show split streams (produced up, shipped down) per category.
        If False and hue is provided, show a separate full-width stream for each (category, hue).
    palette : str or sequence
        Seaborn palette name or list of colors.
    global_scale : bool
        Scale widths across all series (True) or individually by hue (False).
    ax : matplotlib.axes.Axes or None
        Axes to draw on. If None, a new figure and axes are created.
    smoothing_method : {"none", "rolling", "lowess"}
        Smoothing approach to apply to each series.
    smoothing_window : int or float
        Window size for rolling average or frac for LOWESS (0-1).

    Returns
    -------
    matplotlib.axes.Axes
    """
    # Create axes if not provided
    if ax is None:
        fig, ax = plt.subplots(figsize=(14, 6))

    # Sort data and build pivot table
    data = data.sort_values(by=x)
    pivot = data.pivot_table(
        index=x,
        columns=[y, hue] if hue else [y],
        values=units_col,
        aggfunc='sum',
        fill_value=0
    )
    pivot.index = pd.to_datetime(pivot.index)

    # Smoothing
    if smoothing_method == "rolling" and smoothing_window > 1:
        pivot = pivot.rolling(window=int(smoothing_window), center=True, min_periods=1).mean()
    elif smoothing_method == "lowess":
        sm_df = pd.DataFrame(index=pivot.index, columns=pivot.columns, dtype=float)
        x_num = (pivot.index - pivot.index.min()).days
        frac = smoothing_window if 0 < smoothing_window <= 1 else 0.3
        for col in pivot.columns:
            sm_df[col] = lowess(pivot[col].values, x_num, frac=frac, return_sorted=False)
        pivot = sm_df.fillna(0)

    # Plotting
    if hue and not split:
        # One full-width stream per (category, hue)
        series_list = [
            (cat, hval)
            for cat in pivot.columns.levels[0]
            for hval in pivot.columns.levels[1]
        ]
        labels = [f"{cat}-{hval}" for cat, hval in series_list]
        hue_levels = list(pivot.columns.levels[1])
        colors = sns.color_palette(palette, n_colors=len(hue_levels))
        max_val = pivot.values.max() if global_scale else None

        for idx, (cat, hval) in enumerate(series_list):
            s = pivot[(cat, hval)]
            if global_scale:
                w = s / max_val * 0.4
            else:
                lm = pivot.xs(hval, axis=1, level=1).values.max()
                w = s / lm * 0.4 if lm > 0 else 0

            baseline = np.full(len(w), idx, dtype=float)
            top = baseline + w.values
            bottom = baseline - w.values

            ax.fill_between(
                pivot.index,
                bottom,
                top,
                color=colors[hue_levels.index(hval)],
                alpha=0.7
            )

        ax.set_yticks(range(len(labels)))
        ax.set_yticklabels(labels)

    else:
        # Split streams per category
        categories = pivot.columns.levels[0]
        hue_levels = list(pivot.columns.levels[1]) if hue else [None]
        colors = sns.color_palette(palette, n_colors=len(hue_levels)) if hue else ['gray']
        side_map = {'produced': 1, 'shipped': -1} if split and hue else {h: 1 for h in hue_levels}
        max_val = pivot.values.max() if global_scale else None

        for ci, cat in enumerate(categories):
            for hi, hval in enumerate(hue_levels):
                s = pivot[(cat, hval)]
                if global_scale:
                    w = s / max_val * 0.4
                else:
                    lm = pivot.xs(hval, axis=1, level=1).values.max()
                    w = s / lm * 0.4 if lm > 0 else 0

                side = side_map[hval]
                base = np.full(len(w), ci, dtype=float)
                ax.fill_between(
                    pivot.index,
                    base,
                    base + side * w.values,
                    color=colors[hi],
                    alpha=0.7,
                    label=f"{hval}" if ci == 0 else ""
                )

        ax.set_yticks(range(len(categories)))
        ax.set_yticklabels(categories)
        if hue:
            ax.legend(title=hue, loc='upper right')

    ax.set_xlabel(x)
    ax.set_title("Stream Chart: Units by Date and Product Type")
    return ax

# Example usage:
fig, ax = plt.subplots(figsize=(14, 6))
stream_chart(
    data=df,
    x='date',
    y='product_type',
    units_col='units',
    hue='type',
    split=True,
    palette='Set2',
    global_scale=False,
    smoothing_method='lowess',
    smoothing_window=0.1,
    ax=ax
)
plt.tight_layout()
plt.show()
```
